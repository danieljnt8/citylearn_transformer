# Trajectory Transformer Data only from RBC

```python
list_metrics = ["carbon_emissions_total","cost_total",
               "electricity_consumption_total","monthly_one_minus_load_factor_average",
               "ramping_average","zero_net_energy"]
```

```python
df_1 = pd.read_csv(r"checkpoints/city_learn/traj_config_v1_seq20_rf_CombinedReward_norm_wrapper/uniform/model_RBCAgent1_timesteps_100000_rf_CombinedReward_seed_28_norm_wrapper/run_2/results_1.csv",index_col = 1)
df_1 = df_1[df_1.level=="district"]
df_1 = df_1.loc[list_metrics]
df_1 = df_1[["value"]]
```

```python
df_1["RBC"] = [0.893,0.826,0.887,0.953,0.924,1.135]
```

```python
df_1.columns = ["TT","RBC"]
```

```python
df_1
```

```python
df_1.plot(kind='bar', figsize=(10, 6))

# Add labels and title
plt.xlabel('Metrics')
plt.ylabel('Value')
plt.title('Performance Comparison')
plt.xticks(rotation=20)  # Rotate x-axis labels if needed
plt.legend(title='Type')
plt.show()
```
