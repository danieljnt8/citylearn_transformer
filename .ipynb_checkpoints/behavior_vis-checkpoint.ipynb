```python
import pandas as pd
from datasets import load_from_disk
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
path_dt = "data_result/DT/train/data_interactions_train_1.pkl"
path_tt = "data_result/TT/train/data_interactions_train_1.pkl"
path_rbc = "behavior_analysis/data_interactions_RBC.pkl"
path_best = "data_interactions/best_dataset_multi.pkl"
```

```python
data_dt = load_from_disk(path_dt)
data_tt = load_from_disk(path_tt)
data_rbc = load_from_disk(path_rbc)
data_best = load_from_disk(path_best)
```

```python
data_rbc
```

```python
data_best
```

```python
df_dt = pd.DataFrame(data_dt["actions"], columns=["act_1","act_2","act_3","act_4","act_5"])
df_tt = pd.DataFrame(data_tt["actions"], columns=["act_1","act_2","act_3","act_4","act_5"])
df_rbc = pd.DataFrame(data_rbc["actions"], columns=["act_1","act_2","act_3","act_4","act_5"])
df_best = pd.DataFrame(data_best["actions"], columns=["act_1","act_2","act_3","act_4","act_5"])
df_rbc = df_rbc.iloc[:-1,:]
```

```python
# Define a function to extract the float from the list
def extract_float(lst):
    return lst[0] if isinstance(lst, list) and len(lst) > 0 else lst

# Apply the function to each element in the DataFrame
df_rbc = df_rbc.applymap(extract_float)
df_best = df_best.applymap(extract_float)
```

```python
building_no = 1
name_ = "act_{}".format(building_no)
```

```python
diff_dt_rbc = df_dt.loc[:720, name_] - df_rbc.loc[:720, name_]
diff_tt_rbc = df_tt.loc[:720, name_] - df_rbc.loc[:720, name_]
```

```python
diff_dt_rbc
```

```python
plt.figure(figsize=(20, 4))

#plt.plot(df_dt["act_1"], label='DT')
plt.plot(abs(diff_dt_rbc), label='DT')
plt.plot(abs(diff_tt_rbc), label='TT')

# Add titles and labels
plt.title('Action Difference from RBC in Building {}'.format(building_no),fontsize=16)
plt.xlabel('Hour',fontsize=14)
plt.ylabel('Action Value',fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0,0.25)
plt.legend()

plt.savefig("plots_SB/action_difference/train_b_{}.png".format(building_no))

# Show the plot
plt.show()

plt.close()
```

## Rewards

```python
plt.figure(figsize=(20, 4))

#plt.plot(df_dt["act_1"], label='DT')
plt.plot(data_rbc["rewards"][1000:1500], label='RBC')
plt.plot(data_best["rewards"][1000:1500], label='Best')

# Add titles and labels
plt.title('Action Difference from RBC (Building {})'.format(building_no))
plt.xlabel('Hour')
plt.ylabel('Difference')
plt.legend()
plt.savefig("plots_SB/action_difference/train_b_{}.png".format(building_no))

# Show the plot
plt.show()
plt.close()
```
